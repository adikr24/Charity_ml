```python
# pd.get dummies for
# workclass
#marital status
# occupation
#relationship
#race
#sex
#native_country _can remove
df_workclass=pd.get_dummies(df['workclass'])
df[[' Federal-gov', ' Local-gov', ' Private', ' Self-emp-inc',' Self-emp-not-inc', ' State-gov', ' Without-pay']]=df_workclass
# when we have n categories it's advisible to keep n-1 dummies
df.drop([' Federal-gov'],inplace=True,axis=1)
df.drop(['workclass','income'],inplace=True,axis=1)

df_marital=pd.get_dummies(df['marital-status'])
df[['Divorced', 'Married-AF-spouse', 'Married-civ-spouse','Married-spouse-absent', 'Never-married', 'Separated', ' Widowed']]=df_marital
df.drop(['Divorced'],axis=1,inplace=True)

df_occupation=pd.get_dummies(df['occupation'])
df[['Adm-clerical', 'Armed-Forces', 'Craft-repair', 'Exec-managerial',
       'Farming-fishing', 'Handlers-cleaners', 'Machine-op-inspct',
       'Other-service', 'Priv-house-serv', 'Prof-specialty',
       'Protective-serv', 'Sales', 'Tech-support', 'Transport-moving']]=df_occupation
df.drop(['Adm-clerical'],axis=1,inplace=True)

df_relationship=pd.get_dummies(df['relationship'])
df[['Husband', 'Not-in-family', 'Other-relative', 'Own-child','Unmarried', 'Wife']]=df_relationship
df.drop(['Husband'],axis=1,inplace=True)

df_race=pd.get_dummies(df['race'])
df[['Amer-Indian-Eskimo', 'Asian-Pac-Islander', 'Black', 'Other','White']]=df_race
df.drop(['Amer-Indian-Eskimo'],axis=1,inplace=True)

df_sex=pd.get_dummies(df['sex'])
df[['Female', 'Male']]=df_sex
df.drop(['Female'],axis=1,inplace=True)

df.drop(['marital-status','occupation','relationship','race','sex'],axis=1,inplace=True)
```

```python
# for education level
np.unique(df['education_level'])
# from statistical analysis it was found that theses categories in basic _edu is insignificant; thus I tried to bin them together
basic_edu_1=[]
hs_grad_1=[]
associate_1=[]
associate_voc_1=[]
bachelors_1=[]
doctorate_1=[]
masters_1=[]
some_college_1=[]
prof_school_1=[]


for i in df['education_level']:
    basic_edu=0
    hs_grad=0
    masters=0
    doctorate=0
    prof_school=0
    some_college=0
    bachelors=0
    associate=0
    associate_voc=0
    if i == ' 10th' or i== ' 11th' or i==' 12th' or i==' 1st-4th' or i== ' 5th-6th' or i == ' 7th-8th' or i==' 9th' or i==' Preschool':
        basic_edu=1
    elif i== ' HS-grad':
        hs_grad=1
    elif i==' Assoc-acdm':
        associate=1
    elif i==' Assoc-voc':
        associate_voc=1
    elif i== ' Bachelors':
        bachelors=1
    elif i== ' Doctorate':
        doctorate=1
    elif i== ' Masters':
        masters=1
    elif i == ' Prof_school':
        prof_school=1
    elif i== ' Some_college':
        some_college=1
        
    basic_edu_1.append(basic_edu)
    hs_grad_1.append(hs_grad)
    associate_1.append(associate)
    associate_voc_1.append(associate_voc)
    bachelors_1.append(bachelors)
    doctorate_1.append(doctorate)
    masters_1.append(masters)
    some_college_1.append(some_college)
    prof_school_1.append(prof_school)
```

```python
df['basic_edu']=basic_edu_1
df['hs_grad']=hs_grad_1
df['associate']=associate_1
df['associate_voc']=associate_voc_1
df['bachelors']=bachelors_1
df['masters']=masters_1
df['doctorate']=doctorate_1
df['some_college']=some_college_1
df['prof_school']=prof_school_1
```

```python
# logistic regression - Logistic regression works well when there are some continuous variables in the dataset and although this
# algorithm does not take into account the linear dependency, However if any linear dependencies are present then logistic regresison
# performs better than several Other machine learning algorithm -source SAS tutorials
# logistic regression could be used over here since the output is binary and dichotomous and logistic regression would yield good predictions
# if they are taken, However in this particular data set they are rejected because the F1_score was too low compared to other models


## knearest_neighbors: Although Kmeans works well when in classification problems, it's a supervised learning algorithm and classifies
# similar points together. In this case it performed poorly and gave a score which was lower than the F1 _score so this model was also
# rejected


## svm - This machine learning algorithm creates a boundary between two points and punishes points that are misclassified,
# the caveat of this model is that it's quite difficult to comprehend and works like a black box (similar to artificial neural network)
# it's widely known to fit complex data and in our case works fine and give a good score in both the training and crossvalidation phase
# . In layman's term - with an increase in a variable there's a new dimension that get's added so
# in situations like these svm has been proven useful because it can handle data with higher number of variables or dimensions and 
# also it creates complex decision boundaries which come in handy when dimensions increases.


##decision trees - decision trees and quite simple to implement and comprehend and have been proven useful when the dimensionality increases
# in this case the decision trees gave a decent score but we have models which outperformed decision trees and random forest
#random forest - a more robust model of decision trees where trees are randomly split and decisions are made by partial data which falls
# into the group. Upon making several predictions from single trees, a voting is done to see which class was mostly predicted by the trees
# and  then that decision is taken, it's definitely better than decision trees and takes care of redundant variables but in our 
# case it didn't work as expected even after performing the grid search, the tree's performance didn't increase.
# adaboost - similar to random forest and it's a collection of weak classifiers which after every iteration punishes the misclassified 
# point a lot and then combines all the weak classifier together to make a decision. 
# adaboost had worked fine with our data and substituting it with SVM would not be a bad decision as the model score is similar to SVM's
# both in validation and training.
```
